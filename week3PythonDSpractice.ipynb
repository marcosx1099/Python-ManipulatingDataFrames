```python
import numpy as np
import pandas as pd
```

```python
#task1 store the data in ascending order of sales display the result
data = pd.read_csv('Practice.csv')


sort = data.sort_values('Sales', ascending=True)
sort
```

```python
data.shape
```

```python
# task2 display all the rows where the sale amount is over >=45k


sales = data[data.Sales >= 45000]
sales

# or 

sales = data[(data['Sales'] >= 45000)]
sales
```

```python
#task3 Filter the data separately for the states of CA, TX, MT and save them in 3 dataframes
#Sort the 3 datframes in ascending order of sale amount and display them

ca = data[data.State == 'California']
print(f'This is the California DataFrame \n',ca.sort_values('Sales', ascending = True))
print()
tx = data[data.State == 'Texas']
print(f'This is the Texas DataFrame \n',tx.sort_values('Sales', ascending = True))
print()
mt = data[data.State == 'Montana']
print(f'This is the Montana DataFrame \n',mt.sort_values('Sales', ascending = True))
```

```python
#task4 Calculate the sum of sales for the states in task3 and display it
print('Total sales in California:', ca.Sales.sum())
print('Total sales in Texas:', tx.Sales.sum())
print('Total sales in Montana:', mt.Sales.sum())
```

```python
#task5 Display the months with the highest and the lowest monthly sale amount for the store in the state of California
low_month = ca.sort_values('Sales').iloc[:1]
print('The Month with the lowest sale was')
low_month
```

```python
high_month = ca.sort_values('Sales',ascending = False).iloc[:1] #takes the whole row
print('The Month with the highest sale was')
high_month
```
