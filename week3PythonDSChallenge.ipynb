```python
import numpy as np
import pandas as pd
data = pd.read_csv('Challenge.csv')
data
```

```python
#task 1 load the data into a Pandas DataFrame. 
# Identify the # of rows and columns and check 
# if there are any null values in the given data


print(data.shape)
print()
print(data.info(verbose=True))

#no null values
```

```python
#task2 Display the list of females who are self 
# employed and have an income of more than 4k

task2 = data[(data.Gender == 'Female') & (data.Self_Employed =='Yes') & (data.ApplicantIncome > 4000)]

print(task2.shape)

task2
```

```python
#task3 identify those male customers who hold a graduate 
#degree and have the loan status as yes
task3 = data[(data.Gender == 'Male') & (data.Education =='Graduate') & (data.Loan_Status == 'Y')]


print(task3.shape)
task3
```

```python
#task4 calculate the number of customers who have the loan status as yes, 
#three or more dependents, and a property in a rural area
#display all the details of such customers
task4 = data[(data.Dependents == '3+') & (data.Property_Area =='Rural') & (data.Loan_Status == 'Y')]

print(task4.shape)
task4[['Dependents','Property_Area','Loan_Status']]
```

```python
#task6 Idenitfy the customers who are not married, are self employed, live in a nonurban area,
#and have their loan status as no. Calculate and display the gender wise count of such customers
task6 = data[(data.Married == 'No') & (data.Self_Employed == 'Yes') & (data.Property_Area == 'Rural') & (data.Loan_Status == 'N')]

#print(task6.shape)

male_count = task6[task6.Gender == 'Male'].shape[0]
female_count = task6[task6.Gender == 'Female'].shape[0]

print()
print(f'There are {male_count} males and {female_count} females that fit all these conditions')

task6
```

```python
#task7 Calculate and dislpay the average income of customers based 
#on each area where their property is located(urban, semiurban,rural)

semiUrban = data[data.Property_Area == 'Semiurban']
urban = data[data.Property_Area == 'Urban']
rural = data[data.Property_Area == 'Rural']

avg_semiUrban = round(semiUrban["ApplicantIncome"].mean(),2)
avg_urban = round(urban["ApplicantIncome"].mean(),2)
avg_rural = round(rural["ApplicantIncome"].mean(),2)

print(f'The average income of the semiurban area is $ {avg_semiUrban}')
print(f'The average income of the urban area is $ {avg_urban}')
print(f'The average income of the rural area is $ {avg_rural}')
```

```python
#task8 List the customers who have a property in a rural or semiurban area, applicant's income > $4,000, 
#coapplicant's income > $1,000, and loan status is 'N'. Count the number of such customers.

task8 = data[((data.Property_Area == 'Rural') | (data.Property_Area == 'Semiurban')) & (data.ApplicantIncome > 4000) & (data.CoapplicantIncome > 1000) & (data.Loan_Status == 'N')]
print(task8.shape)
task8
```

```python
#task9 Display the average income and the count of customers who have 3+ dependents, 
# live in an urban area, and have their loan status as 'Y'

task9 = data[(data.Dependents == '3+') & (data.Property_Area == 'Urban') & (data.Loan_Status == 'Y')]
task9['ApplicantIncome'].mean()
```

```python
#task10 display the list of customers who live in rural or semiurban areas with 3+ 
#dependents and have an income that is more than the average income 
#of the customers who live in an urban area

task10 = data[(data.Dependents == '3+') & ((data.Property_Area == 'Rural') | (data.Property_Area == 'Semiurban'))]
avg = task10["ApplicantIncome"].mean()

avgTask10 = task10[task10.ApplicantIncome > avg]
avgTask10


```
